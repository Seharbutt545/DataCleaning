Dealing with empty rows
Identifying all the columns where datatype is integer or float; so that missing values could be handel by mean values

```python
int_flt_columns = df.select_dtypes(include=['int', 'float']).columns
int_flt_columns
```

Defining a function to deal with missing values

```python
 def missing_values(col_list):
  for column in col_list:
    if column in df.columns:
        if df[column].isnull().any():
            column_mean = df[column].mean()
            df[column].fillna(column_mean, inplace=True)
            print(f"Filled null values in '{column}' with mean value: {column_mean}")
        else:
            print(f"No null values found in '{column}'")
    else:
        print(f"'{column}' not found in the DataFrame")

# Display the DataFrame after filling null values
print("\nDataFrame after filling null values:")
print(df)



```

identifying column with object datatype consisting null values


```python

object_columns_with_nulls = df.select_dtypes(include=['object']).columns[df.select_dtypes(include=['object']).isnull().any()]
object_columns_with_nulls
```

Filling all null values with their respective mode

```python
for column in object_columns_with_nulls:
    mode_value = df[column].mode()[0]  # Get the mode value
    df[column].fillna(mode_value, inplace=True)
```
